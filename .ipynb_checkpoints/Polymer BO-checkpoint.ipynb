# Fit BO model
### MixedSingleTaskGP:
Perform surrogate in a mixed design space via a specialised kernel

```python
# model=SingleTaskGP(X_train_BO, Y_train_BO)
# mll=ExactMarginalLogLikelihood(model.likelihood, model)
# fit_gpytorch_mll(mll)
# model = fully_bayesian.SaasFullyBayesianSingleTaskGP(X_train_BO, Y_train_BO)
# fit_fully_bayesian_model_nuts(model)

model=MixedSingleTaskGP(X_train_BO,Y_train_BO,cat_dims=[5,6,7,8])
mll=ExactMarginalLogLikelihood(model.likelihood, model)
fit_gpytorch_mll(mll)
```

# Acquisition Function
### EI

```python
from botorch.sampling.normal import SobolQMCNormalSampler
# Sobol sequences use a base of two to form successively finer uniform partitions of the unit interval and then reorder the coordinates in each dimension. (Wiki)

EI = qExpectedImprovement(
    model = model,
    best_f = best_y,
    sampler = SobolQMCNormalSampler(1024)
)
```

### UCB

```python
UCB = qUpperConfidenceBound(
    model = model,
    beta = 0.6,
)
```

## Optimiser
### optimize_acqf_mixed
1. Fixed_features_list can be considered as the bound for categorical features
2. Bounds can be considered as the bound for numerical features
3. This optimiser scan through all possibilities in the fixed_feature list and find the best combination. It then fixed the combination and optimise the numerical features.

```python
# Bounds for each feature, obtained from domain expertise and initial dataset.
bounds = torch.tensor([[-5., 0., 0., 0., 0., 0., 0., 0.,0.],[5., 1., 1., 1., 1., 1., 1., 1., 1.]])
```

```python
# Obtain all possible combinations from categorical features, this will be passed onto fixed_feature_list
import itertools
from collections import ChainMap
from itertools import combinations
fixed_feature_list = [[{5:0},{5:1}],
                      [{6:0},{6:1}],
                      [{7:0},{7:1}],
                      [{8:0},{8:1},{8:2},{8:3}]]
combination = [p for p in itertools.product(*fixed_feature_list)]
combination
```

```python
for x in range(0, len(combination)):
    combination[x] = dict(ChainMap(*combination[x]))
combination
```

```python
X_candidates,_ = optimize_acqf_mixed(
    acq_function = EI, # Acquisition function
    bounds = bounds, # Bounds for numerical features
    q = 5, # Number of suggested candidates, change here for different batch size
    num_restarts = 20,
    raw_samples = 64,
    equality_constraints = [(torch.tensor([1, 2, 3, 4]), torch.tensor([1.]*4), 1.0)], # percentage sum of the compositions
    fixed_features_list = combination # All possible combinations of monomer species
)
X_candidates
```

```python
pd.DataFrame(torch.round(X_candidates,decimals=2))
```

### Substitute generated candidates to random forest classifier

```python
Y_rf = rf.predict_proba(X_candidates)
Y_rf
```

```python
np.count_nonzero(Y_rf[:,1] > 0.5)
```

```python
max(Y_rf[:,1]) # Max. probability achieved
```

```python
# Fix the syntax
Y_rf = torch.tensor(Y_rf[:, 1])
Y_rf = torch.reshape(Y_rf,(len(Y_rf),1))
Y_rf
```

### Second Iteration

```python
X_train_2 = (X_train_BO,X_candidates)
X_train_2 = torch.cat(X_train_2, dim = 0)
X_train_2 = X_train_2.float()
X_train_2
```

```python
Y_train_2 = (Y_train_BO,Y_rf)
Y_train_2 = torch.cat(Y_train_2, dim = 0)
Y_train_2 = Y_train_2.float()
Y_train_2
```

```python
model=MixedSingleTaskGP(X_train_2, Y_train_2,cat_dims=[4,5,6,7])
mll=ExactMarginalLogLikelihood(model.likelihood, model)
fit_gpytorch_mll(mll)
```

```python
EI = qExpectedImprovement(
    model = model,
    best_f = best_y,
    sampler = SobolQMCNormalSampler(1024)
)

UCB = qUpperConfidenceBound(
    model = model,
    beta = 0.6,
)
```

```python
X_candidates,_ = optimize_acqf_mixed(
    acq_function = EI,
    bounds = bounds,
    q = 5,
    num_restarts = 20,
    raw_samples = 64,
    equality_constraints = [(torch.tensor([1, 2, 3, 4]), torch.tensor([1.]*4), 1.0)],
    fixed_features_list = combination
)
X_candidates
```

### Third Iteration

```python
X_train_3 = (X_train_2, X_candidates)
X_train_3 = torch.cat(X_train_3, dim=0)
X_train_3 = X_train_3.float()
X_train_3
```

```python
Y_train_3 = (Y_train_2, Y_rf)
Y_train_3 = torch.cat(Y_train_3, dim=0)
Y_train_3 = Y_train_3.float()
Y_train_3
```

```python
model = MixedSingleTaskGP(X_train_3, Y_train_3, cat_dims=[4, 5, 6, 7])
mll = ExactMarginalLogLikelihood(model.likelihood, model)
fit_gpytorch_mll(mll)
```

```python
EI = qExpectedImprovement(
    model=model,
    best_f=max(Y_train_3),
    sampler=SobolQMCNormalSampler(1024)
)


X_candidates, _ = optimize_acqf_mixed(
    acq_function=EI,
    bounds=bounds,
    q=5,
    num_restarts=20,
    raw_samples=64,
    equality_constraints=[(torch.tensor([1, 2, 3, 4]), torch.tensor([1.] * 4), 1.0)],
    fixed_features_list=combination
)
X_candidates
```
